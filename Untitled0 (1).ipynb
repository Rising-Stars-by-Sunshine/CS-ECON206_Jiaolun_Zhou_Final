From the three charts above, it is easy to see that even though the way SPNE chooses would result in a significant reduction in the global gain for the two, SPNE would still choose this option. However, if we use pasture equilibrium, we will find that we should choose to raise the right number of sheep and cattle in either case, and this is the difference between pasture equilibrium and Nash equilibrium.
The relevance of this thesis is that if we apply the solution strategy of SPNE to the actual society, we will inevitably fall into internal consumption and meaningless waste of resources. Perhaps from the perspective of the game, these differences are only numerical, but in the actual society, these differences imply the inequitable distribution of resources and the irrational use of them.
