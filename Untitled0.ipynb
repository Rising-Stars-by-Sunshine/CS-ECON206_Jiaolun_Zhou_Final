Normal Game Part

```python
import nashpy as nash
import numpy as np

# Creater the game with the payoff matrix

A = np.array([[4, 0],
              [6, 2]]) # A is a company in the market

B = np.array([[4, 6],
              [0, 2]]) # B is a company in the market
     
```

```python


# Form the game
game2 = nash.Game(A,B)
game2
     
```

```python


# Find the Nash Equilibrium with Support Enumeration
equilibria = game2.support_enumeration()
for eq in equilibria:
    print(eq)
```

Describe the Game in text cells.

The name of this game is Competition for Market Share, from Dynamic Models of Oligopoly.

In this matrix, player 1 is Company A and Player 2 is Company B. Every company has two possible options for action: keep the status quo or raise prices. Each cell in the matrix represents the utility of two companies when they choose different strategies.

The game has the following outcomes:
(1) Player A and Player B both keep the same price and get 4
(2) Player A maintains the original price and player B raises the price, player A gets 6 and player B gets 0
(3) If player B maintains the original price and Player A raises the price, player A gets 0 and player B gets 6
(4) Both Player B and Player A raise the price and both get 2

This is similar to the prisoner's dilemma, in which two prisoners turn on each other, leading to a trade-off between maximizing profits and competing for market share.

Elaborate on the Nash equilibrium solution in text cells


If we analyze the problem mathematically, we will have the following process:
First, observe the first column of the matrix, the profit of company A when it chooses to maintain the original price is greater than the profit when it chooses to raise the price, and the primary strategy R is the dominant strategy and can be eliminated. We can use this reasoning to eliminate the second column matrix.
Then, we find that in the second row, the profit of company B choosing to maintain the original price is also greater than the profit of raising the price. So we can eliminate the second row of our matrix.
And then we'll see that the Nash equilibrium in this case is a pure strategy Nash equilibrium where they both choose to keep the original price and they both get 4.
So I think Nashpy is wrong in its conclusion that both companies should have chosen to raise prices. However, this is contradictory to my previous mathematical calculation method, and in this choice method, both of them have to face the risk of 0 return.

Describe this game

This game is the ultimatum game. In this game, one player first suggests a strategy for assigning 2, such as (2, 0), (1,1), or (0, 2). The second player then decides whether or not he agrees with the allocation strategy. If the second player agrees to this allocation strategy, then the money will be allocated according to the allocation proposed by Player A. If the second player disagrees with the distribution, then both players get zero.

We can use Nash equilibrium to analyze this game. However, in the actual experiment, 70 percent of the second player chose to reject the unfair allocation. Even if it means they may not gain anything, they are willing to pay some price to punish the first player for unfair distribution.
At the same time, in the actual experiment, the game is a multi-round game, so some of the first players will choose a more unfair allocation method to retaliate against the second player because the second player rejected his allocation suggestion.
Some second players also choose to accept unfair allocations to please the first player, because they believe that the first player may choose a fairer allocation in return for their compromise.

Elaborate on the solutions

For a computer program, the second player should always choose to agree to the first player's allocation, because if the second player chooses not to accept the first player's allocation, neither player will gain. However, if the second player chooses the first allocation, at least one player gains.
This is why the computer program always chooses to accept the assignment plan in the SPNE.


However, this line of reasoning is based on the rational economic man hypothesis, that is, the solution that always seeks to maximize benefits, and I would like to point out here the difference between this analysis and reality.
In a real experiment, in the cases where the money was not split 50/50,  the proposer's proposal was responded to only 37 percent of the time (Lowe). In the second week of the experiment,  the proposers became more stingy, and in this more unequal distribution,  the proposal acceptance rate dropped to 32 percent (Lowe),  which is inconsistent with the hypothesis of rational economic man.
The reason for this is that I think rational economic man theory excludes changes in participants based on emotional factors or concerns about fairness. Like the example, I presented on Monday.
